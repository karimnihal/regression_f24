```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

```

```python
#Question 5
kbbdata=pd.read_csv('../data/KelleyBlueBookData.csv')
kbbdata.sample(10)
```

```python
kbb =kbbdata[['Price','Mileage','Liter','Cylinder']].copy()
kbb.info()
```

```python
model = smf.ols('Price~Mileage+Liter+Cylinder',data=kbb).fit()
model.summary()
```

```python
sm.stats.anova_lm(model,typ=1)
```

```python
null_model = smf.ols('Price ~ Mileage + Liter', data=kbb).fit()

SSE_H0 = sum(null_model.resid ** 2)
df_SSE_H0 = null_model.df_resid

SSE_H1 = sum(model.resid ** 2)
df_SSE_H1 = model.df_resid

print(f"SSE_H0:{SSE_H0}")
print(f"df_SSE_H0: {df_SSE_H0}")
print(f"SSE_H1: {SSE_H1}")
print(f"df_SSE_H1: {df_SSE_H1}")
```

```python
F_alt = ((SSE_H0 - SSE_H1) / (df_SSE_H0 - df_SSE_H1)) / (SSE_H1 / df_SSE_H1)
print(f"F_alt: {F_alt}")
```

```python
sm.stats.anova_lm(model,typ=2)
```

```python
null_model = smf.ols('Price ~ Liter + Cylinder', data=kbb).fit()

SSE_H0 = sum(null_model.resid ** 2)
df_SSE_H0 = null_model.df_resid

SSE_H1 = sum(model.resid ** 2)
df_SSE_H1 = model.df_resid

print(f"SSE_H0:{SSE_H0}")
print(f"df_SSE_H0: {df_SSE_H0}")
print(f"SSE_H1: {SSE_H1}")
print(f"df_SSE_H1: {df_SSE_H1}")
```

```python
F_alt = ((SSE_H0 - SSE_H1) / (df_SSE_H0 - df_SSE_H1)) / (SSE_H1 / df_SSE_H1)
print(f"F_alt: {F_alt}")
```

```python
#Question 6
model_1 = smf.ols('Price ~ Mileage + Cylinder', data=kbb).fit()
model_1.summary()
```

```python
SSE = sum(model_1.resid ** 2)
SST = sum((kbb['Price'] - kbb['Price'].mean()) ** 2)
n = len(kbb)
p = len(model_1.params)

R2 = 1 - (SSE / SST)
R2_adj = 1 - ((SSE / (n - p)) / (SST / (n - 1)))

print(f"SSE: {SSE}")
print(f"SST: {SST}")
print(f"n: {n}")
print(f"p: {p}")
print(f"R^2: {R2}")
print(f"R^2_adj: {R2_adj}")
```
