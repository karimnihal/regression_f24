```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

## Examples of Heteroskedasticity 

```python
df = pd.read_csv('../Data/kc_house_data.csv')
df['log_price'] = np.log(df['price'])
```

```python
f = 'price~sqft_living'
model2 = smf.ols(formula=f, data=df).fit()
p2 = model2.fittedvalues
res2 = model2.resid
plt.scatter(p2,res2)
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Fitted Values vs. Residuals")
```

```python
f1 = 'log_price~sqft_living'
model3 = smf.ols(formula=f1, data=df).fit()
p3 = model3.fittedvalues
res3 = model3.resid
plt.scatter(p3,res3)
plt.xlabel("Fitted Values")
plt.ylabel("Residual")
plt.title("Fitted Values vs. Residuals")

#the heteroskedasticity is improved, but  this data comes with more problems such as influential points
```

```python
model3.summary()
```

<b>Example of solution: robust standard errors </b>
    
While it is common to use robust standard errors in economics, the documentation in Python’s StatsModels is not as rich as expected. We can grab the robust standard errors, or Heteroskedasticity-Consistent Standard Errors (HCE) using some built-in functions.

```python
credit = pd.read_csv('../Data/Credit.csv')

#orginal model that we know having heteroskedasticity problem
model =smf.ols('Balance~Income + Rating +Cards+Age +Education +Gender+Student+Married+Ethnicity',data=credit).fit()

#extact HCE variance-coviance matrix
V_HC_1 =model.cov_HC0
print(V_HC_1)
```

```python
#sometimes we may not care to see the corrected SE values themselves, 
#we can extract the corrected t test and p values directly if we just care about inference results
print(np.round(model.get_robustcov_results(cov_type = "HC0").summary2().tables[1], 5))
```

```python
#Compare to the orginal model summary, numbers are slightly changed.
model.summary()
```

```python
# weighted least squares: optional
plt.figure(figsize=(20, 6))

# i: index
for i, col in enumerate(credit.columns):
    plt.subplot(2, 6, i+1)
    x = credit[col]
    y = model.resid
    plt.plot(x, y, 'o')

    plt.tight_layout()
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('residuals')
    
    # looks like residuals are proportional to Limit, Balance, and Rating
    
    
```

```python
credit["res_sq"] = model.resid**2
model_res = smf.ols("res_sq ~ Rating", data=credit).fit()  
weight = model_res.fittedvalues
weight = abs(weight) 
weight = weight**-1 
credit['weight'] = weight

```

```python
y=credit.Balance
x=credit[['Income', 'Rating','Cards','Age','Education']] 

```

```python
mod_wls = sm.WLS(y,x.astype(float), credit['weight'])
mod_res = mod_wls.fit()
print(mod_res.summary())
```
