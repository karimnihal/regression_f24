# A Puzzle with Intervals (CIs vs. PIs)

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from plotnine import *
import scipy.stats as stats
```

### First let's review the idea of confidence.

```python
# Generate sample data
np.random.seed(1234321)
sample_size = 100
population_mean = 50
population_std = 10

# Calculate confidence intervals for multiple simulations
num_simulations = 100
confidence_level = 0.95
intervals = []

for i in range(num_simulations):
    sample = np.random.normal(population_mean, population_std, sample_size)
    sample_mean = np.mean(sample)
    # Use ddof=1 for sample standard deviation
    sample_std = np.std(sample, ddof=1)
    margin_of_error = stats.t.ppf(
        (1 + confidence_level) / 2, df=sample_size - 1) * sample_std / np.sqrt(sample_size)
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    contains_mean = (lower_bound <= population_mean) and (
        population_mean <= upper_bound)
    color = 'Success' if contains_mean else 'Miss'
    intervals.append((i, sample_mean, lower_bound, upper_bound, color))


# Convert to DataFrame for plotnine
df_intervals = pd.DataFrame(
    intervals, columns=['Simulation', 'SampleMean', 'Lower', 'Upper', 'Color'])


p = (ggplot(df_intervals, aes(y='Simulation'))
     + geom_segment(aes(x='Lower', xend='Upper',
                    yend='Simulation'), color='blue', alpha=0.2)
     + geom_point(aes(x='SampleMean'), color='blue', size=1)
     + geom_vline(xintercept=population_mean, color='red', linetype='--')
     + annotate("text", y=num_simulations-1, x=population_mean +
                1.05, label='Population Mean', color='red')
     + labs(y='Simulation', x='Confidence Interval',
            title='Confidence Interval Visualization')
     + theme_minimal()
     + theme(
    # Set panel background color
    panel_background=element_rect(fill='lightyellow', color=None),
    # Set the background for the whole plot
    plot_background=element_rect(fill='lightyellow', color=None)
))

p
```

```python
# Plot the confidence intervals using plotnine
pp = (ggplot(df_intervals, aes(y='Simulation'))
      + geom_segment(aes(x='Lower', xend='Upper',
                     yend='Simulation', color='Color'), alpha=0.7)
      + geom_point(aes(x='SampleMean', color='Color'), size=1)
      + geom_vline(xintercept=population_mean, color='black', linetype='--')
      + labs(y='Simulation', x='Confidence Interval',
             title='Confidence Interval Visualization')
      + theme_minimal()
      + theme(
          # Set panel background color
          panel_background=element_rect(fill='lightyellow', color=None),
          # Set the background for the whole plot
          plot_background=element_rect(fill='lightyellow', color=None)
))

pp
```

### CIs in Simple Linear Regression

So far we've learned about CIs for our LS coefficients $\hat{\beta}_k,~k=0,1,$ and also for the mean of the response at given levels of the predictor $E(Y|X=x)$.

They take the form: 

$$\widehat{\beta}_0 \pm t^*_{n-2}(1-\alpha/2)\cdot  \widehat{\sigma}\sqrt{\frac{1}{n} + \frac{\bar{x}^2}{SSX}}$$

$$\widehat{\beta}_1 \pm t^*_{n-2}(1-\alpha/2) \cdot \widehat{\sigma}\sqrt{\frac{1}{SSX}}$$
and 

$$\widehat{y} (x) \pm t^*_{n-2}(1-\alpha/2) \cdot \widehat{\sigma}\sqrt{\frac{1}{n} + \frac{(x_i - \bar{x})^2}{SSX}}.$$
Also recall that the last term $\widehat{y} (x)$ is an estimate of $g(x) = E(y|X = x)$, the mean of the response at a given $x$ value. 

Notice the general form of each of these intervals is:
$$estimate \pm MoE  $$
$$estimate \pm crit. value \cdot (typical~ prediction~ error)$$
$$estimate \pm crit. value \cdot \widehat{SE}(estimate - target)$$

(MoE = margin of error)

```python
true_beta_0 = 2  # True intercept
true_beta_1 = 5  # True slope
sigma = 10  # Standard deviation of errors

# Generate sample data parameters
sample_size = 100
num_simulations = 100
confidence_level = 0.95

# Store results
results = []

for i in range(num_simulations):
    # Simulate the data
    X = np.random.uniform(0, 10, sample_size)
    errors = np.random.normal(0, sigma, sample_size)
    Y = true_beta_0 + true_beta_1 * X + errors

    # Add constant for intercept
    X_with_const = sm.add_constant(X)

    # Fit the linear regression model
    model = sm.OLS(Y, X_with_const).fit()

    # Get the estimates for the intercept and slope
    beta_0_hat = model.params[0]
    beta_1_hat = model.params[1]

    # Get the standard errors for the estimates
    beta_0_se = model.bse[0]
    beta_1_se = model.bse[1]

    # Calculate the t-critical value
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 2)

    # Calculate the confidence intervals for beta_0 and beta_1
    beta_0_margin_of_error = t_critical * beta_0_se
    beta_1_margin_of_error = t_critical * beta_1_se

    beta_0_lower = beta_0_hat - beta_0_margin_of_error
    beta_0_upper = beta_0_hat + beta_0_margin_of_error
    beta_1_lower = beta_1_hat - beta_1_margin_of_error
    beta_1_upper = beta_1_hat + beta_1_margin_of_error

    # Check if the true values are within the confidence intervals
    beta_0_contains_true = (beta_0_lower <= true_beta_0 <= beta_0_upper)
    beta_1_contains_true = (beta_1_lower <= true_beta_1 <= beta_1_upper)

    # Color coding: Success if CI contains the true value, Miss otherwise
    beta_0_color = 'Success' if beta_0_contains_true else 'Miss'
    beta_1_color = 'Success' if beta_1_contains_true else 'Miss'
    
    # Store the results
    results.append({
        'Simulation': i,
        'Beta_0_Hat': beta_0_hat,
        'Beta_0_Lower': beta_0_lower,
        'Beta_0_Upper': beta_0_upper,
        'Beta_0_Contains': beta_0_contains_true,
        'Beta_0_Color': beta_0_color,
        'Beta_1_Hat': beta_1_hat,
        'Beta_1_Lower': beta_1_lower,
        'Beta_1_Upper': beta_1_upper,
        'Beta_1_Contains': beta_1_contains_true,
        'Beta_1_Color': beta_1_color
    })

# Convert to DataFrame for plotting
df_results = pd.DataFrame(results)

# Plot the confidence intervals for the intercept with color and background theme
p1 = (ggplot(df_results, aes(y='Simulation'))
      + geom_segment(aes(x='Beta_0_Lower', xend='Beta_0_Upper', yend='Simulation', color='Beta_0_Color'), alpha=0.7)
      + geom_point(aes(x='Beta_0_Hat', color='Beta_0_Color'), size=1)
      + geom_vline(xintercept=true_beta_0, color='red', linetype='--')
      + scale_color_manual(values={'Success': 'lightblue', 'Miss': 'red'})
      + labs(y='Simulation', x='Confidence Interval for Intercept', title='Confidence Interval for Intercept (\u03B2\u2080)')
      + theme_minimal()
      + theme(
          panel_background=element_rect(fill='lightyellow'),
          plot_background=element_rect(fill='lightyellow')
      )
)

# Plot the confidence intervals for the slope with color and background theme
p2 = (ggplot(df_results, aes(y='Simulation'))
      + geom_segment(aes(x='Beta_1_Lower', xend='Beta_1_Upper', yend='Simulation', color='Beta_1_Color'), alpha=0.7)
      + geom_point(aes(x='Beta_1_Hat', color='Beta_1_Color'), size=1)
      + geom_vline(xintercept=true_beta_1, color='red', linetype='--')
      + scale_color_manual(values={'Success': 'blue', 'Miss': 'red'})
      + labs(y='Simulation', x='Confidence Interval for Slope', title='Confidence Interval for Slope (\u03B2\u2081)')
      + theme_minimal()
      + theme(
          panel_background=element_rect(fill='lightyellow'),
          plot_background=element_rect(fill='lightyellow')
      )
)

# Display the plots
p1, p2
```

### Now let's explore on a real dataset. 

A criminologist studied the relationship between level of education and crime rate. He collected data from 84 medium-sized US counties. Two variables were measured: X  the percentage of individuals having at least a high-school diploma; and Y  the crime rate (crimes reported per 100, 000 residents) in the previous year. The complete data is in crimerate.txt.


```python
# Load the crime rate data
crimerate = pd.read_csv("../data/crimerate.txt", sep='\s+') #uses \s+ to match one or more whitespace characters (spaces, tabs, etc.).

# Number of observations
n = len(crimerate)
crimerate.columns = ['rate', 'pct_diploma']

crimerate.head()
```

```python
mod_crime = smf.ols('rate ~ pct_diploma', data=crimerate).fit()

# Summary of the model
print(mod_crime.summary())
```

```python
# 1. The fitted value when x = 80
beta_hat = mod_crime.params

def yhat(x):
    return beta_hat[0] + beta_hat[1] * x
```

```python
# 2. Critical value
alpha = 0.05
crit = stats.t.ppf(1 - alpha/2, df=n-2)
```

$SE (\hat{\theta}) = \sqrt{Var(\hat{\theta})}$

```python
# 3. Estimated standard error of E(Y|X = x)
sigma2hat = sum(mod_crime.resid**2) / (n - 2)
sigmahat = np.sqrt(sigma2hat)
x_values = crimerate['pct_diploma']
xbar = np.mean(x_values)
SSX = sum((x_values - xbar)**2)

def seyhat(x):
    return sigmahat * np.sqrt(1/n + (x - xbar)**2 / SSX)
```

```python
# Standard error at x = 80
seyhat_value = seyhat(x=80)
print(f"Standard error at x = 80: {seyhat_value.round(4)}")
```

```python
# 95% CI for the mean crime rate when HS grad rate is 80%
ci = yhat(x=80) + np.array([-1, 1]) * crit * seyhat(x=80)
print(f"95% CI: ", ci)

# Simulate a new county with a HS grad rate of 80% and check if it falls within our interval.
# Technically we dont know the true parameters, so it's hard to simulate this. 
# We just have to use our best guesses for the parameters based on our estimates and pretend they are the ground truth here... 
beta_0 = 20517.6 
beta_1 = -170.5752 
sigma = 2356.292

np.random.seed(123)
x = 80

def ynew(x):
    eps = np.random.normal(0, sigma, 1)
    return beta_0 + beta_1 * x + eps


ynew_value = ynew(x=80)[0]
print(f"My simulated new observation has a rate of {ynew_value.round(4)} when pct_diploma = 80.")
```

### Is that value captured in my CI?

```python
captured = (ynew_value > ci[0]) & (ynew_value < ci[1])
print("New value captured in CI:", captured)
```

#### Hmm... 
The new value is not in the interval, but that can happen with CIs.  Maybe we were just unlucky-- let's try to simulate 10,000 counties with HS grad rates of 80\% and see if we get the proper coverage of 95\%.

```python
# Simulate N = 10,000 counties' crime rates with HS grad rates of 80%
N = 10000
np.random.seed(123)
eps = np.random.normal(0, sigma, N)
ynew_values = beta_0 + beta_1 * x + eps

captured_values = (ynew_values > ci[0]) & (ynew_values < ci[1])
coverage_rate = np.sum(captured_values) / N
print("Coverage rate:", coverage_rate)
```

Not even close... what's going wrong???
